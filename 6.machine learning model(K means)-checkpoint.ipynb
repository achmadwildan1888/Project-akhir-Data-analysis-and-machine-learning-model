```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
df = pd.read_csv('data.csv')
df.columns[28:54]
df = df.drop(['Unnamed: 0','ID','Photo','Flag','Club Logo','Joined','Special','Loaned From','Body Type', 'Release Clause',
               'Weight','Height','Contract Valid Until','Wage','Value'], axis = 1)
df = df.dropna()

pd.set_option('display.max_columns', None)  # or 1000
pd.set_option('display.max_rows', None)  # or 1000
pd.set_option('display.max_colwidth', -1)  # or 199

```

```python
df.head()
```

```python
from sklearn.cluster import KMeans
```

```python
#elbow method

ssr=[]
for x in range(1,11):
    model = KMeans(n_clusters=x)
    model.fit(df[['Jersey Number','Overall']])
    ssr.append(model.inertia_)    
```

```python
#elbow method
plt.plot(
        np.arange(1,11),
        ssr,'b-')
plt.xlabel('range')
plt.ylabel('sumsquare error')
plt.grid(True)
```
