Melakukan konversi kolom value dan wage

```python
def konversi(a):
    try:
        value = float(a[1:-1])
        suffix = a[-1:]

        if suffix == 'M':
            value = value * 1000000
        elif suffix == 'K':
            value = value * 1000
    except ValueError:
        value = 0
    return value
```

```python
df['Value'] = df['Value'].apply(konversi)
df['Wage'] = df['Wage'].apply(konversi)


df.Wage.mean()
```

```python
#dengan cara diatas
fitur = ['Value','Wage']

i=0
while i < len(fitur):
    print('Best {0} : {1}'.format(fitur[i],df.loc[df[fitur[i]].astype('float64').idxmax()][1]))
    i += 1

    
```
