```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
df = pd.read_csv('data.csv')
df.columns[28:54]
df = df.drop(['Unnamed: 0','ID','Photo','Flag','Club Logo','Joined','Special','Loaned From','Body Type', 'Release Clause',
               'Weight','Height','Contract Valid Until','Wage','Value'], axis = 1)
df = df.dropna()

pd.set_option('display.max_columns', None)  # or 1000
pd.set_option('display.max_rows', None)  # or 1000
pd.set_option('display.max_colwidth', -1)  # or 199

```

```python
df.head()
```

random forest melihat data secara chronological dengan tipe katagorical . misal jika si A kerja di bank X , dan si B kerja di bank X apakah pola2 seperti itu dapat menghasilkan suatu pola gaji (tipenya kategorical) 

dari random forest saya ingin membuat model untuk reputasi apakah pemain hebat (overall) , dan berpotential bisa menjadi model yang tepat untuk international reputation : 

```python
df[['Overall','Potential','International Reputation']].head()
```

```python
X = df[['Overall','Potential']]
y= df['International Reputation']  #targetnya 
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
```


# Menggunakan Decission Tree# 

```python
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier()
dtree.fit(X_train,y_train)
```

```python
predictions = dtree.predict(X_test)
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,predictions))
```

```python
print(dtree.score(X_train,y_train)*100,'%')
```

```python
prediksi =dtree.predict([X_test.iloc[0]])
aktual = y_test.iloc[0]
print('Hasil prediksi : ',prediksi)
print('Hasil aktual:',aktual) 
```

# Menggunakan RandomForest


```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=600)
rfc.fit(X_train,y_train)
```

```python
predictions = rfc.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,predictions))
print(rfc.score(X_train,y_train)*100,'%')
```

Final result : hasil sama , karena dua algoritma ini secara hierarchial sama maka runtutan algoritmanya kemungkinan sama

# Membuat fitur baru terhadap rating pemain

Case selanjutnya saya ingin membuat model apakah hanya dengan melihat umur dapat membuat model yang baik untuk menentukan pemain tersebut hebat atau tidak?

```python
df['rating pemain']="kosong"
```

```python
df.tail()
```

# Bikin fitur dari overall menjadi categorical

```python

conditions = [
    (df['Overall'] > 90),
    (df['Overall'] > 80),
    (df['Overall'] > 70),
    (df['Overall'] > 60)
]
choices = ['superb', 'excelent', 'good',' not bad']
df['rating pemain'] = np.select(conditions, choices, default='normal')
```

```python
df.tail()
```

```python
from sklearn.preprocessing import LabelEncoder
labelrating = LabelEncoder()
df['ratingLE'] = labelrating.fit_transform(df['rating pemain']) 
```

```python
df.tail()
```

# modeling apakah age adalah acuan yg tepat untuk overall


```python
#disini target labelnya adalah overall yang suda dikategorikan, dan akan learning dari age terhadap overall ...

X = df[['Age']]
#X = df[['Age']]
y= df['ratingLE']  #targetnya 
```

```python
df[['Age','ratingLE']].head(10)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier()
dtree.fit(X_train,y_train)
```

```python
print(dtree.score(X_train,y_train)*100,'%')
```

```python
prediksiage=dtree.predict([X.iloc[0]])
print(prediksiage)
print(y_test.iloc[0])
```
